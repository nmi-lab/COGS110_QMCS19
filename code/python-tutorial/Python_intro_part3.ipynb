#### Introduction

Visualization is one of the most important things of data analysis. `matplotlib` is a powerful package for ___Python___ users. Let's start with an example.

First we generate sample data

```python
X_arr = np.array(range(10))
X_arr
```

```python
Y_arr = 3*X_arr + 2 + np.random.random(X_arr.size) # linear with some noise
Y_arr
```

```python
print(X_arr)
print(Y_arr)
```

To plot a simple scatter plot, we can use `plt.scatter()` function

```python
plt.scatter(X_arr, Y_arr)
plt.plot(X_arr,Y_arr)
```

This is still too simple - we do not even have X or Y axis labels. Further we may want to change the scatter markers. Let's get a better plot!

```python
# Use `+` as marker; color set as `g` (green); s is the marker size
```

```python
plt.scatter(X_arr, Y_arr, marker='|', c='k', s=1000) 
plt.plot(X_arr, Y_arr, linestyle='dashed', color='k')
plt.axis([0, 10, 0, 50])
plt.xlabel('Y')
plt.ylabel('My Y Axis')
plt.title('My First Plot')
```

The following plots a histogram (we'll see histograms later in this class, but for now you only need to understand that it is a ways of representing the distribution of random variables)

```python
gaussian_arr = np.random.normal(0,1, size=1000)
```

```python
plt.hist(gaussian_arr);
```

We can actually customize and make it prettier

```python
# Facecolor set to green; opacity (`alpha`) level: 30%
plt.hist(gaussian_arr, facecolor='g', alpha=0.5)
# Add grid
plt.grid()
```
